# Identifying Malicious Nodes in IoT Networks using Bayes' Theorem

```python
from IoTgraph import IoTgraph
```

The `IoTgraph` class provides an environment to simulate various algorithms that aim to identify the malicious nodes in the network. The environment is defined with the following hyperparameters:
* `network_size` - the number of devices in the network.
* `edge_prob` - how likely it is that any two devices in the network are connected 
* `percent_mal` - the percentage of malicious devices in the network 
* `attack_probs` - how likely it is that a malicious device will attack on a single transmission; each device's individual attack probability is drawn from a uniform distribution using this list as bounds

An environment can be created in one of two ways:
1. In-notebook, by specifying all of the hyperparameters.
2. From a file in the `env` folder (see `env/env_format.txt` for instructions on creating an env file).
Examples of both ways of generating an environment are shown below. 

```python
graph = IoTgraph(fname='./env/env1.txt')
# graph = IoTgraph(network_size=50, edge_prob=0.1, percent_mal=0.3, attack_probs=[0.2, 0.8])
graph.render()
```
