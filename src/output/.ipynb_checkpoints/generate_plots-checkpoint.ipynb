```python
import csv
import matplotlib.pyplot as plt
```

```python
nodes = []
acc = []
prec = []
rec = []

with open('accuracy_results.txt', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    next(reader)
    
    rows = [[int(row[0]), float(row[1]), float(row[2]), float(row[3])] for row in reader if row]
    
    for row in rows:
        nodes.append(row[0])
        acc.append(row[1])
        prec.append(row[2])
        rec.append(row[3])
```

```python
plt.scatter(nodes, prec)
plt.xlabel('Number of nodes in network')
plt.ylabel('Precision')
plt.show()
```

```python
plt.scatter(nodes, rec)
plt.xlabel('Number of nodes in network')
plt.ylabel('Recall')
plt.show()
```
